### Problèmatique.
Un étang abrite une population de poissons dont on veut estimer le nombre $N$. Comme il est impossible de recenser tous les individus de l'étang on utilise une méthode appelée *capture - marquage - recapture*.

Cette méthode consiste à capturer un échantillon de taille $N_0$ de cette population et à marquer les individus. Les individus marqués sont relachés dans l'étang. Après un certain délai permettant aux individus marqués de bien se mélanger à la population de l'étang on procède à un nouveau prélévement de taille $N_1$ et on compte le nombre d'individus marqués : $N_2$.

Nous allons simuler cette situation grace à un programme Python.

La population de poissons d'un étang a été générée aléatoirement. Le nombre de poissons est inconnu. Au départ tous les poissons sont non marqués (bleus).
