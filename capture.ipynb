## Intervalle de confiance.
Lorsqu'on a observé une fréquence $f$ de poissons marqués dans un prélèvement de taille $n$ on appelle intervalle de confiance à 95% l'intervalle $[f - \frac{1}{\sqrt{n}} ; f - \frac{1}{\sqrt{n}} ]$. On considère que cet intervalle contient la vraie valeur de la fréquence avec une sécurité de 95%.
